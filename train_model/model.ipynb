```python
import pandas as pd
import numpy as np
```

# Clean data

```python
df = pd.read_csv('UCI_Credit_Card.csv')
df = df.drop(columns=['ID','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6'])
df = df.rename(columns={"default.payment.next.month": "Default"})
df.head(5)
```

```python
# re_values columns
sex_condition = [df['SEX']==1,df['SEX']==2]
sex_values = ['Male','Female']
df['Sex'] = np.select(sex_condition, sex_values)

edu_condition = [df['EDUCATION']==1,df['EDUCATION']==2,df['EDUCATION']==3,
                (df['EDUCATION']==4)|(df['EDUCATION']==0)|(df['EDUCATION']==5)|(df['EDUCATION']==6)]
edu_values = ['Graduate','Undergraduate','HighSchool','Others']
df['Edu'] = np.select(edu_condition, edu_values)

marry_condition = [df['MARRIAGE']==1,df['MARRIAGE']==2,
                (df['MARRIAGE']==3)|(df['MARRIAGE']==0)]
marry_values = ['Married','Single','Others']
df['Marriage'] = np.select(marry_condition, marry_values)

df = df.drop(columns=['SEX','EDUCATION','MARRIAGE'])

# get dummies
encoder_df = pd.get_dummies(df, columns = ['Sex','Edu','Marriage'])
encoder_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scale= MinMaxScaler()
# standardization of dependent variables
scaled_data = scale.fit_transform(encoder_df.drop(columns=['Default'])) 
# x values 
X_data = pd.DataFrame(scaled_data)
X_data.columns = encoder_df.drop(columns=['Default']).columns
X_data
```

```python
X_data.columns
```

```python
Y_data = encoder_df['Default']
```

# Train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, Y_data, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)

y_pred  =  classifier.predict(X_test)

from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test,y_pred)

print(ac)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)

y_pred  =  classifier.predict(X_test)

from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test,y_pred)

print(cm)
print(ac)
```
